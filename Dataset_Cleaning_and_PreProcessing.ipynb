## 4. Genre Column Transformation (Major Step)

### Problem:

* `listed_in` had combined values like:

  ```
  Action, Adventure, Comedy
  ```

### Solution:

* Split genres into individual values
* Cleaned spaces and inconsistencies

```python

```

## 5. One-Hot Encoding for Genre

* Converted multi-label genres into binary columns

```python

genre_dummies = genre_dummies.add_prefix('Genre_')

```

* Removed duplicate genre columns
